```python
import random

def simulate_once():
    price = 120.00
    target = 130.00
    steps = 2160
    for _ in range(steps):
        move = random.choices(population=[+0.01, 0.0, -0.01],weights=[0.1, 0.85, 0.05],k=1)[0]
        price += move
        price = round(price, 2)
        if price >= target:
            return True

    return False
trials = 100000
successes = sum(simulate_once() for _ in range(trials))
probability = successes / trials
print(f"Estimated Probability: {probability:.4f}")
```
